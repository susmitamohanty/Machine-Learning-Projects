```python
def compute_loss(y,y_pred):
        loss = -np.mean(y*(np.log(y_pred)) - (1-y)*np.log(1-y_pred))
        return loss
```

```python
class LR():
    def __init__(self,X,y,lr= 0.0001,epochs=1000):
        self.lr=lr
        self.epochs = epochs
        self.X = X
        self.y = y 
        self.w=np.zeros(self.X.shape[1])
        self.b=0
        self.losses = []
        
    def sigmoid(self,z):
        p = 1/(1+np.exp(-z))
        return p
    
    def compute_loss(self,y,y_pred):        
        loss=(-self.y * np.log(y_pred) - (1 - self.y) * np.log(1 - y_pred)).mean()
        return loss        
   
    def fit(self):
        #losses =[]
        
        for i in range(self.epochs):
            #self.X is m by n and self.w is 1 by n
            
            z= self.X.dot(self.w) + self.b
            y_pred = self.sigmoid(z)
            error = (y_pred-self.y)   
            
            loss =  self.compute_loss(self.y,y_pred)
            self.losses.append(loss)
            
            # compute gradients
            dw = (1*(self.X.T.dot(error)))/self.X.shape[0] #X is (80,4), error is (80,), hence X.T
            db = (1*np.sum(error))/self.X.shape[0]
            
            #update weight and bias with gradients
            self.w = self.w - (self.lr * dw)
            self.b = self.b - (self.lr * db)
    
    def predict(self,X):
        z = X.dot(self.w) + self.b
        y_pred =1/(1+np.exp(-z))
        y_pred_prob = np.where(y_pred>0.5,1,0)    
     
        return np.asarray(y_pred_prob)      
       
```

```python
lr_model=LR(X_train_sc,y_train)
```

```python
lr_model.fit()
```

```python
y_pred = lr_model.predict(X_test_sc)
y_pred
```

```python
def metrics(y,y_pred):
    tp,tn,fp,fn = 0,0,0,0
    for i in range(len(y)):
        if y[i]==1 and y_pred[i]==1:
            tp +=1
        if y[i]==1 and y_pred[i]==0:
            fn +=1
        if y[i]==0 and y_pred[i]==1:
            fp +=1
        if y[i]==0 and y_pred[i]==0:
            tn +=1
            
    precision = tp/(tp+fp)
    print("Precision",precision)
    recall = tp/(tp+fn)
    print("recall",recall)
    f1_score = 2*precision*recall/(precision+recall)
    print("f1_score",f1_score)
    return precision,recall,f1_score
        
        
```
