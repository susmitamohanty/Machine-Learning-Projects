### Outliers

Outliers are observations that are far away from the other data points in a random sample of a population.

- Extreme values, however, can have a significant impact on conclusions drawn from data or machine learning models.
- When identified, outliers may reveal unexpected knowledge about a population.
- These measurements often do not belong to the population we are interested in and therefore need treatment


Types of outliers

- Univariate outliers(extreme values in the distribution of a specific variable)
    - human age measurement of 120 years
- multivariate outliers(combination of values in an observation that is unlikely)
    - observation of a human with a height measurement of 2 meters (in the 95th percentile) and a weight measurement of 50kg (in the 5th percentile)

###### Handling outliers using Z-score method

- Z score is also called standard score. This score helps to understand if a data value is greater or smaller than mean and how far away it is from the mean. More specifically, Z score tells how many standard deviations away a data point is from the mean.

    - Z-score = (data_point-mean)/Standard Deviation

- A normal distribution is shown below and it is estimated that
    - 68% of the data points lie between +/- 1 standard deviation.
    - 95% of the data points lie between +/- 2 standard deviation
    - 99.7% of the data points lie between +/- 3 standard deviation

- Generally a threshold value of 3 is used as cut-off. If the z score of a data point is more than 3, it indicates that the data point is quite different from the other data points. Such a data point can be an outlier.

###### Trimming

There are 4 methods of handling any outliers detection technique - Trimming, Capping, treating as missing value and Discretization.

Here we will discuss only Trimming and Capping.

- Method of dropping the outliers is called Trimming.
- Excludes the outlier values from our analysis.
- By applying this technique our data becomes thin when there are more outliers
  present in the dataset.
- Its main advantage is its fastest nature.

```python
import numpy as np
from scipy import stats

def zscore_method(df,var):
    #find boundary values
    min_value = df[var].mean()-3*df[var].std() 
    max_value = df[var].mean()+3*df[var].std() 
    print("Boundary values:",[min_value,max_value])
    
    #find outliers
    outliers_df = df[(df[var] < min_value) | (df[var] > max_value) ] 
    
    #remove the outliers
    new_df = df[(df[var] >= min_value) & (df[var] <= max_value) ]
    
    return outliers_df, new_df
    
    
df1,df2  = zscore_method(df,'CRIM')
print(df.shape)
print(df1.shape)
print(df2.shape)
```

###### Capping 

- Method of dropping the outliers is called Capping
- In this technique, we cap our outliers data and make the limit i.e,
  above a particular value or less than that value, all the values will 
  be considered as outliers, and the number of outliers in the dataset gives that capping number.

```python
def zscore_method_capping(df,var):
    #find boundary values
    min_value = df[var].mean()-3*df[var].std() 
    max_value = df[var].mean()+3*df[var].std() 
    print("Boundary values:",[min_value,max_value])
    
    #find outliers
    outliers_df = df[(df[var] < min_value) | (df[var] > max_value) ] 
    
    #remove the outliers
    new_df = df
    #capping
    new_df[var] = np.where(new_df[var] > max_value,max_value, np.where(new_df[var]<min_value, min_value,new_df[var]))
    
    return outliers_df, new_df

df1,df2  = zscore_method_capping(df,'CRIM')
print(df.shape)
print(df1.shape)
print(df2.shape)
```

###### Handling Outliers using IQR method

- IQR is used to measure variability by dividing a data set into quartiles. The data is sorted in ascending order and split into 4 equal parts. Q1, Q2, Q3 called first, second and third quartiles are the values which separate the 4 equal parts.

    - Q1 represents the 25th percentile of the data.
    - Q2 represents the 50th percentile of the data.
    - Q3 represents the 75th percentile of the data.

- If a dataset has 2n / 2n+1 data points, then
    - Q1 = median of the dataset.
    - Q2 = median of n smallest data points.
    - Q3 = median of n highest data points.

- IQR is the range between the first and the third quartiles namely Q1 and Q3: IQR = Q3 – Q1. The data points which fall below Q1 – 1.5 IQR or above Q3 + 1.5 IQR are outliers.
- For skewed distributions use IQR technique.

```python
#IQR method by trimming
def IQR_method(df,var):
    percentile25 = df[var].quantile(0.25)
    percentile75 = df[var].quantile(0.75)
    
    iqr = percentile75 - percentile25
    
    min_value = percentile25-1.5*iqr
    max_value = percentile75+1.5*iqr
    
    outliers_df = df[(df[var] > max_value) | (df[var] < min_value)]
    
    new_df = df[(df[var] >= min_value) & (df[var] <= max_value) ]
    
    return outliers_df, new_df 
        
df1,df2=IQR_method(df,'CRIM')
print(df.shape)
print(df1.shape)
print(df2.shape)    
```

```python
#IQR method by capping
def IQR_method_capping(df,var):
    percentile25 = df[var].quantile(0.25)
    percentile75 = df[var].quantile(0.75)
    
    iqr = percentile75 - percentile25
    
    min_value = percentile25-1.5*iqr
    max_value = percentile75+1.5*iqr
    
    outliers_df = df[(df[var] > max_value) | (df[var] < min_value)]
    
    new_df = df
    #capping
    new_df[var] = np.where(new_df[var] > max_value,max_value, np.where(new_df[var]<min_value, min_value,new_df[var]))
    
    return outliers_df, new_df 
        
df1,df2=IQR_method_capping(df,'CRIM')
print(df.shape)
print(df1.shape)
print(df2.shape)    
```

###### Handling outliers using Percentile method

- This technique works by setting a particular threshold value, which decides based on our problem statement.
- While we remove the outliers using capping, then that particular method is known as Winsorization.
- Here we always maintain symmetry on both sides means if remove 1% from the right then in the left

```python
import numpy as np
from scipy import stats

def percentile_method(df,var):
    #find boundary values
    min_value = df[var].quantile(0.01) 
    max_value = df[var].quantile(0.99) 
    print("Boundary values:",[min_value,max_value])
    
    #find outliers
    outliers_df = df[(df[var] < min_value) | (df[var] > max_value) ] 
    
    #remove the outliers
    new_df = df[(df[var] >= min_value) & (df[var] <= max_value) ]
    
    return outliers_df, new_df
    
    
df1,df2  = percentile_method(df,'CRIM')
print(df.shape)
print(df1.shape)
print(df2.shape)
```

```python
#Percentile method by capping
def Percentile_method_capping(df,var):
    min_value = df[var].quantile(0.01) 
    max_value = df[var].quantile(0.99) 
    
    outliers_df = df[(df[var] > max_value) | (df[var] < min_value)]
    
    new_df = df
    #capping
    new_df[var] = np.where(new_df[var] > max_value,max_value, np.where(new_df[var]<min_value, min_value,new_df[var]))
    
    return outliers_df, new_df 
        
df1,df2=IQR_method_capping(df,'CRIM')
print(df.shape)      
print(df1.shape)
print(df2.shape)  
```

### Multivariate Outliers

- For multivariate outliers, distance metrics is used. The distance between two vectors is determined and can be between two different observations (rows) or an observation (row) and the mean vector (row of means of all columns). 
- Distance metrics can be calculated independent of the number of variables in the dataset (columns).
- The most widely used distance metrics is Mahalanobis distance (MD), that measures the distance between a data point x and a distribution D. It is a multivariate generalization of the internally studentized residuals (z-score). This means the MD defines how many standard deviations x is away from the mean of D.

It canbe defined as below.

-  Here, x represents an observation vector and µ the arithmetic mean vector of independent variables (columns) in the sample. C(-1) is the inverse covariance matrix of the independent variables in the sample.

- The  MD of each observation is compared to a cut-off point. Assuming a multivariate normal distribution of the data with K variables, the Mahalanobis distance follows a chi-squared distribution with K degrees of freedom. 

```python
import scipy as sp
from scipy.stats import chi2

def mahalanobis_method(d):

    x_diff = df - np.mean(d)
    
    #Covariance
    covariance = np.cov(d.values.T)  
    
    #Inverse covariance
    inv_covariance = sp.linalg.inv(covariance)     
    
    #dot product
    left = np.dot(x_diff, inv_covariance) 
    
    mahal = np.dot(left, x_diff.T)
    mahalanobis_distance = mahal.diagonal() 
        
    return mahalanobis_distance
```

```python
import copy
```

```python
#creating a new column in the dataframe that holds
#the Mahalanobis distance for each row
df_copy = copy.deepcopy(df)
df_copy['Mahalanobis']=mahalanobis_method(df_copy)
df_copy.head()
#df.head()
```

```python
#calculate p-value that corresponds to the Chi-Square statistic of the 
#Mahalanobis distance with k-1 degrees of freedom, where k = number of variables.
df_copy['p-value'] = 1 - chi2.cdf(df_copy['Mahalanobis'], 3)
df_copy.head()
```

```python
outliers_df = df_copy[df_copy['p-value']<0.001]
outliers_df.head()
```

The observations having a p-value less than 0.001 is assumed to be an outlier. In this example, the above outliers are selected.

###### Handling outliers using KNN

- kNN is a supervised algorithm that uses 𝑘 neighbours of a data point as basis for classification. The
only parameters required are 𝑘, the amount of neighbours to consider. For example, when classifying a point 𝑝 with 𝑘 = 5, the 5 nearest neighbours of 𝑝 are considered.By looking at the classes of the 𝑘 neighbours, the majority class is chosen as 𝑝’s class.

- In anomaly detection, kNN can be used to detect global anomalies by finding the knearest-neighbors. The anomaly score of a data point can be calculated in a few different ways. One option is to use the average distances to its k-nearest-neighbors. This approach has the benefit of also taking into account the local density of the points 

- It is easy to implement and is a lazy learning algorithm and therefore doesn't require training on all data points

```python
df_copy2 = copy.deepcopy(df)
df_copy2.head()
```

###### Handling outliers using Isolation forest

- IsolationForest is an unsupervised model, in which randomly sub-sampled data is processed in a tree structure based on randomly selected features
- The samples that travel deeper into the tree are less likely to be anomalies as they required more cuts to isolate them. 
- Similarly, the samples which end up in shorter branches indicate anomalies as it was easier for the tree to separate them from other observations.
- Isolation Forests are nothing but an ensemble of binary decision trees. And each tree in an Isolation Forest is called an Isolation Tree(iTree). The algorithm starts with the training of the data, by generating Isolation Trees.Step-by-step, isolationForest works as below.
    - When given a dataset, a random sub-sample of the data is selected and assigned to a binary tree.
    - Branching of the tree starts by selecting a random feature (from the set of all N features) first. And then branching is done on a random threshold ( any value in the range of minimum and maximum values of the selected feature).
    - If the value of a data point is less than the selected threshold, it goes to the left branch else to the right. And thus a node is split into left and right branches.
    - This process from branching is continued recursively till each data point is completely isolated or till max depth(if defined) is reached.
    - The above steps are repeated to construct random binary trees.
- Isolation Forests are computationally efficient and have been proven to be very effective in Anomaly/Outlier detection
- The final anomaly score depends on the contamination parameter, provided while training the model. This implies that we should have an idea of what percentage of the data is anomalous beforehand to get a better prediction.
- Also, the model suffers from a bias due to the way the branching takes place.


```python
from sklearn.ensemble import IsolationForest


random_state = np.random.RandomState(42)

iso=IsolationForest(n_estimators=100,
                      max_samples='auto',
                      contamination=float(0.2),
                      random_state=random_state)
iso
```

#### Handling outliers with LOF 

- Local Outlier Factor (LOF) algorithm is an unsupervised anomaly detection method which computes the local density deviation of a given data point with respect to its neighbors.
- It considers as outliers the samples that have a substantially lower density than their neighbors. This example shows how to use LOF for novelty detection. When LOF is used for outlier detection, we MUST not use predict, decision_function and score_samples on the training set as this would lead to wrong results. 
- We must only use these methods on new unseen data (which are not in the training set). See User Guide : for details on the difference between outlier detection and novelty detection and how to use LOF for outlier detection.
- The number of neighbors considered, (parameter n_neighbors) is typically set
    - 1) greater than the minimum number of samples a cluster has to contain, so that other samples can be local outliers relative to this cluster, and 
    - 2) smaller than the maximum number of close by samples that can potentially be local outliers. 
- In practice, such informations are generally not available, and taking n_neighbors=20 appears to work well in general


```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors=20, contamination=.03)
lof
```

```python
#fit the model with x dataset and get the prediction data with the fit_predict() method.
y_pred = lof.fit_predict(X)
y_pred
```

```python
#extract the negative outputs as the outliers.
lofs_index = np.where(y_pred==-1)
outliers = X[lofs_index]
```
